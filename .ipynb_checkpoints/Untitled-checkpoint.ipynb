```python
import numpy as np
import matplotlib.pylab as plt
from math import *
from scipy import stats
import traceback

nth = 8 #--- number of thresholds
nsamp = 8 #--- number of samples
DIR='./multipleNth'

for ith in xrange(nth):
    nij_shuffled[ith] = {}
    nij_declustered[ith] = {}
    for isamp in xrange(nsamp):
        try:
            nij_shuffled[ith][isamp]=np.load('%s/Run%s/nij_shuffled.npy'%(DIR+'%s'%ith,isamp))
            nij_declustered[ith][isamp] = np.load('%s/Run%s/nij_declustered.npy'%(DIR+'%s'%ith,isamp))
        except:
            traceback.print_exc()
            continue        
```

```python
ncol=6
nrow = int(np.ceil(1.0*nsamp / ncol))
fig= plt.figure(figsize=(ncol*2.0,nrow*2.0))
#
pval = {} #np.zeros(nsamp)
for ib in xrange(nth):
    pval[ib] = {}
    for isamp in xrange(nsamp):
        try:
            #--- p(nij) for declustered
            nmin=nij_declustered[ib][isamp].min()
            nmax=nij_declustered[ib][isamp].max()
            bins=np.linspace(nmin,nmax,int(nmax-nmin)*8)
            hist, edge = np.histogram(nij_declustered[ib][isamp],bins=bins,normed=True)

            #--- p(nij) for shuffled
            nmin=nij_shuffled[ib][isamp].min()
            nmax=nij_shuffled[ib][isamp].max()
            bins=np.linspace(nmin,nmax,int(nmax-nmin)*8)
            hist_shuffled, edge_shuffled = np.histogram(nij_shuffled[ib][isamp],bins=bins,normed=True)
            
            #--- pvalue
            pval[ib][isamp] = stats.ks_2samp(nij_shuffled[ib][isamp],nij_declustered[ib][isamp]).pvalue
        
            if ib != 5:
                continue

            #------------
            #--- plot
            #------------    
        #    fig= plt.figure(figsize=(4,4))
            ax = fig.add_subplot(nrow,ncol,isamp+1)
            ax.set_ylabel('$P(n_{ij})$')
            ax.set_xlabel('$log n_{ij}$')
            ax.set_xlim(floor(nmin),ceil(nmax))
            ax.set_xlim(-5,6)
            ax.set_ylim(1e-3,1)
            #ax.set_xscale('log')
            ax.set_yscale('log')
            ax.plot(edge[:-1],hist,'-',color='black')
            ax.plot(edge_shuffled[:-1],hist_shuffled,
                    '-',color='red',
                   label= '%3.2e'%(pval[ib][isamp]))
            ax.legend()
        except:
            traceback.print_exc()
            continue
```

```python
fig=plt.figure(figsize=(6,6))
ax=fig.add_subplot(111)
#ax.set_xlim(0,1.0)
ax.set_ylim(1e-10,1)
#ax.legend()
ax.set_xlabel('$n_{th}$')
ax.set_ylabel('$p$')
ax.set_yscale('log')

for ib in xrange(nth):
    if ib in pval:
        for isamp in xrange(nsamp):
            if isamp in pval[ib]:
                ax.scatter(ib,pval[ib][isamp],marker='x',color='black')

                
```
